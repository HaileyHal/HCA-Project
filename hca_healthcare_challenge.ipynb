```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

hurricanes = pd.read_excel('/content/atlantic hurricane data.xlsx')
hurricanes = hurricanes.drop(columns=['time(hrs/mins)','34ktwindNE', '34ktwindSE', '34ktwindSW','34ktwindNW', '50ktwindNE', '50ktwindSE', '50ktwindSW', '50ktwindNW','64ktwindNE', '64ktwindSE', '64ktwindSW', '64ktwindNW'])
# change L in made landfall to y, fill in null of made landfall to n
hurricanes['made landfall'] = hurricanes['made landfall'].apply(lambda x: 'Y' if x == ' L' else 'N')
#add in slashes into date(year/month/day)
def add_slashes_to_date(date):
    date_str = str(date).zfill(8)  # Convert to string and pad with zeros if needed
    return f"{date_str[:4]}/{date_str[4:6]}/{date_str[6:]}"
hurricanes['Date'] = hurricanes['Date(year/month/day)'].apply(add_slashes_to_date)
#drop all rows with a date that start with A
hurricanes = hurricanes[~hurricanes['Date'].str.startswith('A')]
#dropping old date column
hurricanes = hurricanes.drop(columns=['Date(year/month/day)'])
hurricanes.head()
```

```python
hurricanes.to_excel('cleaned_hurricanes_data.xlsx', index=False)
from google.colab import files
files.download('cleaned_hurricanes_data.xlsx')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

airfare = pd.read_csv('/content/Consumer_Airfare_Report__Table_1_-_Top_1_000_Contiguous_State_City-Pair_Markets_20250220.csv')
# drop columns
airfare = airfare.drop(columns=['quarter', 'carrier_lg', 'large_ms', 'carrier_low', 'lf_ms', 'table_1_flag', 'Geocoded_City1', 'Geocoded_City2'])
airfare = airfare.drop(columns=['citymarketid_1', 'citymarketid_2'])

#delete non/metropolitan area from city 1 and 2
airfare['city1'] = airfare['city1'].str.removesuffix('(Metropolitan Area)')
airfare['city2'] = airfare['city2'].str.removesuffix('(Metropolitan Area)')
airfare['city1'] = airfare['city1'].str.removesuffix('(Nonmetropolitan)')
airfare['city2'] = airfare['city2'].str.removesuffix('(Nonmetropolitan)')

# separating state from city column and creating new state columns
airfare['state1'] = airfare['city1'].str.split(',').str[1].str.strip()
airfare['state2'] = airfare['city2'].str.split(',').str[1].str.strip()
airfare['state1'] = airfare['state1'].str.split(',').str[0].str.strip()
airfare['state2'] = airfare['state2'].str.split(',').str[0].str.strip()

# getting rid of state in city column
airfare['city1'] = airfare['city1'].str[:-4]
airfare['city2'] = airfare['city2'].str[:-4]
airfare.head()
```

```python
airfare.to_excel('cleaned_airfare_data.xlsx', index=False)
from google.colab import files
files.download('cleaned_airfare_data.xlsx')
```
