```python
import pandas as pd
import numpy as np

hurricane = pd.read_csv('/content/storms.csv.zip')
hurricane.head()
```

```python
hurricane.info()
```

```python
hurricane['category'].value_counts()
```

```python
# delete unnamed column
hurricane = hurricane.drop(columns=['Unnamed: 0', 'hour', 'tropicalstorm_force_diameter', 'hurricane_force_diameter', 'wind', 'pressure', 'lat', 'long'])
# fill in nans in category with 0
hurricane['category'] = hurricane['category'].fillna(0)
hurricane.head()
```

```python
# create a new column, date, by combining year, month, day
hurricane['date'] = hurricane['year'].astype(str) + '/' + hurricane['month'].astype(str) + '/' + hurricane['day'].astype(str)
hurricane.head()
```

```python
# how many duplicate records do I have?
hurricane.duplicated().sum()
```

```python
# I am losing some granularity, but I don't want it. the only duplicate hurricanes at this point will be ones that spanned multiple days
# delete duplicate records
hurricane = hurricane.drop_duplicates()
hurricane.head()
```

```python
# drop all rows where status is tropical depression
hurricane = hurricane[hurricane['status'] != 'tropical depression']
hurricane = hurricane[hurricane['status'] != 'tropical storm']
hurricane = hurricane[hurricane['status'] != 'extratropical']
hurricane.head()
```

```python
hurricane = hurricane.drop(columns=['status'])
hurricane.head()
```

```python
# I have some records that are identical except for category. How can I keep the one with the highest value for category and delete the rest?
hurricane = hurricane.sort_values('category', ascending=False).groupby('name').first().reset_index()
hurricane.head()
```
