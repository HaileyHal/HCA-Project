```python
import pandas as pd
import numpy as np
AK = pd.read_csv('/content/NRI_Table_Counties_Alaska.csv')
AR = pd.read_csv('/content/NRI_Table_Counties_Arkansas.csv')
CA = pd.read_csv('/content/NRI_Table_Counties_California.csv')
CO = pd.read_csv('/content/NRI_Table_Counties_Colorado.csv')
FL = pd.read_csv('/content/NRI_Table_Counties_Florida.csv')
GA = pd.read_csv('/content/NRI_Table_Counties_Georgia.csv')
ID = pd.read_csv('/content/NRI_Table_Counties_Idaho.csv')
IN = pd.read_csv('/content/NRI_Table_Counties_Indiana.csv')
KA = pd.read_csv('/content/NRI_Table_Counties_Kansas.csv')
KY = pd.read_csv('/content/NRI_Table_Counties_Kentucky.csv')
LA = pd.read_csv('/content/NRI_Table_Counties_Louisiana.csv')
MO = pd.read_csv('/content/NRI_Table_Counties_Missouri.csv')
NV = pd.read_csv('/content/NRI_Table_Counties_Nevada.csv')
NH = pd.read_csv('/content/NRI_Table_Counties_NewHampshire.csv')
NC = pd.read_csv('/content/NRI_Table_Counties_NorthCarolina.csv')
SC = pd.read_csv('/content/NRI_Table_Counties_SouthCarolina.csv')
TN = pd.read_csv('/content/NRI_Table_Counties_Tennessee.csv')
TX = pd.read_csv('/content/NRI_Table_Counties_Texas.csv')
UT = pd.read_csv('/content/NRI_Table_Counties_Utah.csv')
VA = pd.read_csv('/content/NRI_Table_Counties_Virginia.csv')
```

```python
# r bind all datasets
df = pd.concat([AK,AR,CA,CO,FL,GA,ID,IN,KA,KY,LA,MO,NV,NH,NC,SC,TN,TX,UT,VA])
df.info()
```

```python
df.head()
```

```python
import pandas as pd
# List of columns to keep
columns_to_keep = ['OID_', 'NRI_ID', 'STATE', 'STATEABBRV', 'COUNTY', 'COUNTYTYPE', 'POPULATION', 'HRCN_EVNTS', 'HRCN_AFREQ', 'HRCN_RISKV', 'RISK_VALUE', 'RISK_SCORE', 'RISK_RATNG']

# Keep only the specified columns
df = df[columns_to_keep]

df.head()
```

```python
# take the log of hrcn risk
df['lg_HRCN_RISKV'] = np.log(df['HRCN_RISKV'])
```

```python
df.head()
```

```python
# fill nans with 0
df['lg_HRCN_RISKV'] = df['lg_HRCN_RISKV'].fillna(0)
df['HRCN_RISKV'] = df['HRCN_RISKV'].fillna(0)
df['HRCN_EVNTS'] = df['HRCN_EVNTS'].fillna(0)
df['HRCN_AFREQ'] = df['HRCN_AFREQ'].fillna(0)
df.head()
```
