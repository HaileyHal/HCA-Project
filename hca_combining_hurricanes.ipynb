```python
import pandas as pd
import numpy as np
hurricane = pd.read_excel('/content/cleaned_NEWhurricane_data.xlsx')
costly = pd.read_excel('/content/cleaned_mostcostlyhurricanedata.xlsx')
```

```python
hurricane.head()
```

```python
# making sure my column names match
costly.rename(columns={'City/Name':'name', 'Year':'year', 'Category':'category'}, inplace = True)
costly.head()
```

```python
merged = hurricane.merge(costly, on = ['name', 'year', 'category'], how='left')
merged.head()
```

```python
merged['date'].value_counts()
```

```python
# show all rows where date is 2016/8/28
merged[merged['date'] == '2020/9/17']
```
